## Q1. 클로저를 사용하여 주어진 숫자 리스트에서 최댓값과 최솟값을 찾는 함수를 작성하세요.
1. find_min과 find_max함수를 반환해야한다.
2. find_min과 find_max는 각각 최솟값과 최댓값을 반환해야한다.
3. min_value와 ma_value를 함수 내 지역변수로 저장해둔뒤 내부함수를 통해 호출해 private하게 한다.

````python
def find_min_max(numbers):
    # 지역 변수로 최솟값과 최댓값을 계산하여 저장
    min_value = min(numbers)
    max_value = max(numbers)

    # 호출시 최솟값을 반환
    def find_min():
        return min_value
    
    # 호출시 최댓값을 반환
    def find_max():
        return max_value
    
    # 최솟값과 최댓값을 반환하는 내부함수를 튜플로 반환
    return find_min, find_max
````

````python
numbers = [10, 5, 8, 12, 3, 7]
find_min, find_max = find_min_max(numbers)

print('최솟값:', find_min())
print('최댓값:', find_max())
````

코드 실행 순서
```
1. 정의된 find_min_max(numbers) 함수를 호출하며 int값으로 구성된 list를 인자로 전달
2. 외부 함수에서 지역 변수인 min_value, max_value값 계산 후 할당
3. 내부 함수인 find_min, find_max를 정의해 지역 변수인 min_value, max_value를 저장해 반환하는 클로저 역할 수행
4. 두 함수를 튜플 형태로 한번에 반환
4. 반환된 함수를 각각 find_min(), find_max() 형태로 호출해 클로저 공간에서 min_value와 max_value를 꺼내어 반환
```

## Q2. 클로저를 활용하여 함수 호출 횟수를 출력하는 데코레이터를 만들어봅시다. 
1. 클로저기 때문에 내부에 데코레이터 함수를 만들고 반환해야한다.
2. 외부 함수에 실행 횟수를 저장할 지역 변수를 만들어둔다.
3. 내부 함수에서 지역 변수를 수정해야하기 때문에 nonlocal을 사용해야한다.

- Case1. 매개변수를 추가해주지 않은 버전

````python
def counter(func):
    # 실행 횟수를 저장해둘 지역 변수
    count = 0
    
    # 데코레이션하는 내부 함수
    def wrapper():
        # 지역 변수 참조
        nonlocal count
        count += 1
        
        func()
        print(f'{func.__name__} 실행횟수 : {count}')
    return wrapper
````

````python
@counter
def say_hello(name):
    print('Hello AIFFEL!', name)

for i in range(5):
    say_hello(i)
````

Result : 함수를 받아와서 데코레이션할 때 함수의 매개변수를 고려하지 않으면 Error가 발생한다.

- Case2. 매개변수를 추가해준 버전

````python
def counter(func):
    # 실행 횟수를 저장해둘 지역 변수
    count = 0

    # 데코레이션하는 내부 함수
    def wrapper(*args, **kwargs):
        # 지역 변수 참조
        nonlocal count
        count += 1
        func(*args, **kwargs)
        print(f'{func.__name__} 실행횟수 : {count}')
    return wrapper
````

````python
@counter
def say_hello(name):
    print('Hello AIFFEL!', name)

for i in range(5):
    say_hello(i)
````

Result2. 매개변수가 있는 함수를 데코레이션해도 정상적으로 작동하는것을 볼 수 있다.

코드 실행 순서
```
1. counter(func) 함수 정의
2. @counter 데코레이터 선언 확인
3. say_hello 함수를 counter(say_hello)로 실행 
4. counter 내부에서 count = 0으로 초기화
5. wrapper 함수를 정의하며 count=0을 클로저로 저장
6. wrapper를 반환하여 앞으로 say_hello는 wrapper로 작동
```
