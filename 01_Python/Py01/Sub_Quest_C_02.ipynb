Q1. 회문 판독 및 분류기

```python
def is_palindrome(target):
    return target == target[::-1]
```

```python
def palindrome_classifier(targets):
    result_dict = {'회문O':[],'회문X':[]}
    for target in targets:
        if is_palindrome(target):
            result_dict['회문O'].append(target)
        else:
            result_dict['회문X'].append(target)
    palindrome_count = len(result_dict['회문O'])
    total_count = len(targets)
    ratio = palindrome_count / total_count if total_count != 0 else 0.0
    return result_dict, palindrome_count, ratio
```

```python
target1 = "아많다많다많다많아"
target2 = "아정말많다많다많아"

print(is_palindrome(target1))
print(is_palindrome(target2))
```

```python
targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']

print(palindrome_classifier(targets))
```

Q2. 성적 분석기

```python
scores = {
    '유원영' : {'국어': 94, '영어': 91, '수학': 89},
    '설지민' : {'국어': 88, '영어': 96, '수학': 90},
    '신윤아' : {'국어': 86, '영어': 88, '수학': 77},
    '장윤아' : {'국어': 98, '영어': 76, '수학': 92}
}
```

```python
subjects = ['국어', '영어', '수학']
student_count = len(scores)
total_scores = {subject : 0 for subject in subjects}
for student in scores:
    for subject in subjects:
        total_scores[subject] += scores[student][subject]
class_avg = {subject : total_scores[subject] / student_count for subject in subjects}
class_total_avg = sum(class_avg.values()) / len(subjects)
```

```python
def score_analysis(student_name):
    if student_name not in scores:
        print(f'{student_name}의 데이터가 없습니다.')
        return
    print(f'[{student_name} 성적 분석]')
    for subject in subjects:
        score = scores[student_name][subject]
        avg = class_avg[subject]
        diff = score - avg
        diff_str = f'+{diff:.2f}' if diff >= 0 else f'{diff:.2f}'
        print(f'{subject}: {score}점, 학급 평균({avg:.2f}점) 대비 {diff_str}')
    
    print()
    student_total_avg = sum(scores[student_name][subject] for subject in subjects) / len(subjects)
    total_diff = student_total_avg - class_total_avg
    total_diff_str = f'+{total_diff:.2f}' if total_diff >= 0 else f'{total_diff:.2f}'
    print(f'전과목 평균: {student_total_avg:.2f}, 학급 평균({class_total_avg:.2f}점) 대비 {total_diff_str}')
```

```python
score_analysis('설지민')
```
