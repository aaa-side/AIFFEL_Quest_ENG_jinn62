```python
import math
```

```python
# 사용자로부터 올바른 정수를 입력받을 때까지 반복하는 함수
# ValueError가 발생 시 오류 메시지를 출력하고 재입력을 반복
def get_int_input(prompt):
    while True:
        try:
            return int(input(prompt))   # 정수값일때(ValueError가 발생하지 않을 때) 반환
        except ValueError: # ValueError발생 시 메시지 출력 후 반복
            print("잘못된 입력입니다. 정수를 입력해주세요.")
```

```python
# 계산기 루프
while True:
    # 1. 정수 입력 받기
    num1 = get_int_input("첫 번째 정수를 입력해주세요.")
    num2 = get_int_input("두 번째 정수를 입력해주세요.")

    # 2. 연산자 입력 받기
    operator = input("연산자를 입력해주세요(+,-,*,/,**)")
    
    # 3. 연산 수행 및 예외 처리
    try:
        if operator == '+':
            ans = num1 + num2
        elif operator == '-':
            ans = num1 - num2
        elif operator == '*':
            ans = num1 * num2
        elif operator == '/':
            ans = num1 / num2   # num2가 0이면 ZeroDivisionError 발생
        elif operator == '**':
            # math,pow를 통해 **계산-
            ans = math.pow(num1, num2)
        else:   # 지원하지 않는 연산자일 경우 예외 처리
            raise Exception("지원하지 않는 연산자입니다.")
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다.")
    # 지원하지 않는 연산자 또는 그외 오류 처리
    except Exception as e:
        print(f'{e}')

    # 예외 처리가 발생하지 않았을 경우 결과 출력
    else:
        print(f'결과: {num1} {operator} {num2} = {ans}')
    
    # 4. 계산기 종료 여부 확인
    if input("계속 계산하시겠습니까? (y/n)") != 'y':
        print("계산기 종료")
        break
```

2. 나눗셈 연산 시 두번째 정수가 0인 경우 ZeroDivisionError를 처리해 메시지를 출력해야한다.
