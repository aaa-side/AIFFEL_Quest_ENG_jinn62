# Project 1. 사각형 넓이 구하기
- 직사각형, 평행사변형, 사다리꼴의 넓이를 모두 구할 수 있는 클래스 구현   
    - 직사각형 : 가로 x 세로   
    - 평행사변형 : 밑변 x 높이   
    - 사다리꼴 : (밑변 + 윗변) x 높이 x 1/2   

### 코드 작성
1. 생성자 : 1, 2, 3이 아니면 메세지 출력 후 while loop반복으로 재입력   
2. rect() : 가로, 세로를 입력받아 넓이 반환   
3. par() : 밑변, 높이를 입력받아 넓이 반환   
4. trape() : 밑변, 윗변, 높이를 입력받아 넓이 반환   

```python
# 사격형 넓이 구하기
class Square:
    # 생성자 : 사각형 종류를 입력받음(예외 처리 진행)
    def __init__(self):
        while True:
            try:
                # 종류 입력 받기
                self.square = int(input("사각형의 모양을 고르시오.\n 1. 직사각형 2. 평행사변형 3. 사다리꼴\n >>>"))
            except:
                # 숫자가 아닐 때
                print("1, 2, 3 중에서 다시 입력하세요.")
            else:
                # 입력 받은 번호에 따라 메시지 출력
                if self.square == 1:
                    print('직사각형 함수는 rect()입니다.')
                    break
                elif self.square == 2:
                    print('평행사변형 함수는 par()입니다.')
                    break
                elif self.square == 3:
                    print('사다리꼴 함수는 trape()입니다.')
                    break
                else:
                    # 1, 2, 3이외의 숫자 처리
                    print('1, 2, 3 중에서 다시 입력해주세요')
    
    # 직사각형 넓이 계산
    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하시오.(ex. 가로, 세로)\n>>>').split(','))
        return f'직사각형의 넓이는 {width * vertical} 입니다.'

    # 평행사변형 넓이 계산
    def par(self):
        bottom, height = map(int, input('밑변, 높이를 입력하시오.(ex. 밑변, 높이)\n>>>').split(','))
        return f'평행사변형의 넓이는 {bottom * height} 입니다.'
    
    # 사다리꼴 넓이 계산
    def trape(self):
        bottom, top, height = map(int, input('밑변, 윗변, 높이를 입력하시오.(ex. 밑변, 윗변, 높이)\n>>>').split(','))
        return f'사다리꼴의 넓이는 {(bottom + top) * height / 2} 입니다.'
```

```python
# 객체 3개 생성(각 생성 시 종류 입력)
square1 = Square()
square2 = Square()
square3 = Square()
```

```python
# 객체별 넓이 함수 실행 및 출력
print(square1.rect())
print(square2.par())
print(square3.trape())
```

### 코드 실행
1. Square 객체 3개를 생성해 0, 1, 2, 3 확인(0일 땐 재입력)  
2. 각 객체별 함수 실행 (입력 : 2, 3 / 3, 2 / 3, 2, 1)    
3. 결과 확인

# Project 2. Kiosk 만들기
- 메뉴판 출력   
- 메뉴 고르기   
- 추가 주문   
- 결제하기   
- 주문표 출력   

### 코드 작성
1. 생성자 : 메뉴 및 가격 설정
2. print_menu() : 메뉴판 출력
3. select_menu() : 메뉴 정하기
4. pay() : 추가 주문 및 취소 후 결제
5. table() : 주문표 출력

```python
import sys

class Kiosk:
    # 생성자: 기본 메뉴판 데이터 설정 및 장바구니 초기화
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []
        self.order_price = []
        self.print_menu()   # 객체 생성 시 메뉴판 자동 출력

    # 메뉴판 출력 함수
    def print_menu(self):
        for i in range(len(self.menu)):
            print(f'{i+1}. {self.menu[i]} : {self.price[i]}원')

    # 메뉴 선택 및 장바구니 관리 함수
    def select_menu(self):
        while True:
            choice = int(input("주문할 음료 번호를 입력하세요.(종료는 0번) : "))

            # choice 분기
            # 0 입력시
            if not choice:
                if not self.order_menu:
                    print("장바구니가 비어있습니다. 키오스크를 종료합니다.")
                    sys.exit()
                else:
                    return self.pay()

            # 범위를 벗어난 번호 입력시
            if not (1 <= choice <= len(self.menu)):
                print("없는 메뉴입니다. 다시 주문하세요.")
                continue

            # 정상 메뉴 번호 선택 시 HOT/ICE 선택 루프
            idx = choice - 1
            while True:
                tmp = int(input("HOT 음료는 1번, ICE 음료는 2번을 입력하세요. : "))
                # 번호에 맞는 메뉴 장바구니 추가
                if tmp == 1:
                    self.order_menu.append('HOT ' + self.menu[idx])
                    break
                elif tmp == 2:
                    self.order_menu.append('ICE ' + self.menu[idx])
                    break
                else:
                    print("올바른 번호(1 or 2)를 입력하세요")
            # 선택한 메뉴 가격 추가
            self.order_price.append(self.price[idx])

            # 장바구니 관리 루프
            while True:
                # 현재 장바구니 내역 출력
                for i in range(len(self.order_menu)):
                    print(f'[{i+1}] {self.order_menu[i]} : {self.order_price[i]}원')
                
                print("1 : 추가 주문 | 2. 메뉴 취소 | 3. 선택 완료")
                next = int(input("다음 단계를 선택하세요."))

                # 1. 추가 주문 : 루프 종료 후 메뉴 선택으로 이동
                if next == 1:
                    break
                # 2. 메뉴 취소 : 장바구니에서 특정 메뉴 취소
                elif next == 2:
                    if not self.order_menu:
                        print("장바구니가 이미 비어있습니다.")
                        break

                    cancel = int(input("취소할 메뉴의 번호를 입력하세요."))

                    if (1<= cancel <= len(self.order_menu)):
                        cancel_idx = cancel - 1
                        removed_menu = self.order_menu.pop(cancel_idx)
                        self.order_price.pop(cancel_idx)
                        print(f'{removed_menu}가 장바구니에서 삭제되었습니다.')
                        continue    # continue를 통해 다시 장바구니 관리로 돌아감
                    else:
                        print("올바른 번호를 입력하세요.")
                    # 3. 선택 완료: 결제 단계 진입 또는 종료
                elif next == 3:
                    if not self.order_menu:
                        print("장바구니가 비어있습니다. 키오스크를 종료합니다.")
                        sys.exit()
                    return self.pay()
                else:
                    print("올바른 번호(1 or 2 or 3)을 입력하세요.")           

    # 결제 수단 선택 및 진행 함수
    def pay(self):
        while True:
            # 총 결제 금액 계산
            self.total_price = sum(self.order_price)
            # for i in range(len(self.order_menu)):
            #     print(f'[{i+1}] {self.order_menu[i]} : {self.order_price[i]}원')
            print("1 : 결제 | 2 : 메뉴 담기")

            choice = int(input("다음 단계를 선택하세요."))
            # 결제 진행 분기
            # 1. 결재 진행
            if choice == 1:
                print("결제 수단을 선택하세요.")
                print("현금('cash' 또는 1) | 카드('card' 또는 2)")
                payment = input("선택 : ")
                if payment in ['1', 'cash', '현금']:
                    print("직원을 호출하겠습니다. 잠시만 기다려 주세요.")
                    self.table()
                    break
                elif payment in ['2', 'card', '카드']:
                    print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                    self.table()
                    break
                else:
                    print("올바른 번호(1 or 2)를 입력하세요.")
            # 2. 메뉴 다시 담기(초기 메뉴로 돌아가기)
            elif choice == 2:
                print("메뉴 화면으로 돌아갑니다.")
                self.select_menu()
                break
            else:
                print("올바른 번호(1 or 2)를 입력하세요.")

    # 영수증 출력 함수
    def table(self):
        print("---------영수증---------")
        for i in range(len(self.order_menu)):
            print(f'{self.order_menu[i]} : {self.order_price[i]}원')
        print(f'총 합계 : {self.total_price}원')
        print("------------------------")


# 키오스크 객체 생성 및 주문 시작
if __name__ == "__main__":
    cafe = Kiosk()
    cafe.select_menu()
```

### 코드 실행
1. cafe = Kiosk()가 실행되면 인스턴스가 생성되며 __init__을 통해 초기 데이터가 설정
- self.menu, self.price : 메뉴 설정 self.order_menu, self.order_price : 빈 리스트로 초기화 self.print_menu() 호출
2. select_menu() : 메뉴 번호 입력   
   2-1. choice를 통해 분기   
      1. 0을 입력한 경우(종료 또는 결제 단계 진입)
      - 장바구니가 비어있다면 sys.exit()으로 프로그램 종료
      - 장바구니에 물건이 있다면 self.pay() 호출
      2. 범위를 벗어난 번호를 입력한 경우
      - continue를 통해 다시 메뉴 입력창으로 복귀
      3. 메뉴 번호를 정상적으로 입력한 경우
      - idx를 계산하고 다음 단계로 진행
   2-2. HOT / ICE 선택 루프
      1. 1 또는 2가 입력될 때까지 루프 반복
      - 각 번호에 맞게 메뉴 앞에 추가해 order_menu에 append
      - 루프 종료 후 order_price에 append
3. 장바구니 관리
- order_menu를 출력하고 다음 단계를 고른다.(1 : 추가 주문 | 2. 메뉴 취소 | 3. 선택 완료)
   1. 추가 주문 선택시
   - break를 통해 탈출 후 메뉴 분기로 돌아간다.
   2. 메뉴 취소 선택시
   - 장바구니가 비어있다면 메뉴 분기로 돌아간다.
   - 장바구니에 물건이 있다면 취소할 메뉴의 번호를 입력받아 idx를 구한다.
   - pop을 통해 물건을 장바구니에서 비우고 다음 단계를 다시 고른다
   3. 선택 완료 선택시
   - 장바구니가 비어있다면 sys.exit()으로 프로그램 종료
   - 장바구니에 물건이 있다면 self.pay() 호출
4. 결제 단계(pay())
   1. 최종 주문을 확인한다. (1 : 결제 | 2 : 메뉴 담기)
   - 메뉴 담기 선택시 다시 select_menu로 돌아간다.
   2. 지불 수단(현금 or 카드)를 입력받는다.
   3. 올바른 값이 입력될 때까지 반복하고 table()을 호출한다.
5. 영수증 출력(table())
- 최종 장바구니를 출력하고 총 금액을 출력한다.
